```python
import pandas as pd
import numpy as np
diabeticdata = pd.read_csv('diabetic_data.csv')
diabeticdata
```

```python
diabeticdata.describe()
```

```python
t1 = pd.DataFrame(diabeticdata, columns =['time_in_hospital','insulin','num_lab_procedures'])
t1
```

```python
t1.dropna(subset=['time_in_hospital'])
t1
```

```python
from scipy import stats
s = s = pd.Series(t1.time_in_hospital)
print(s)
s = stats.zscore(s)
print('mean: ', np.mean(s))
print('std: ', np.std(s))
```

```python
t1['normalized_time_in_hospital'] = s
t1
```

```python
ohe = pd.get_dummies(t1['insulin'], prefix = 'Insulin')
print(ohe)
# add the one-hot-encoding (ohe) to the dataframe (df2) using pandas concate method
t1 = pd.concat([t1, ohe], axis = 1)
t1
```

```python
# The size of bins are determined such that equal
# data sample are divided into equal sized bins.
# This is done using Quantile. For example, if 
# there are 5 bins, the dataset is divided into 
# five quatiles. Each quatile containing 20% of  
# total area of distribution.

# Defining a function  binning for reusability
def binning(data, numOfBin, col):
    data[col] = pd.cut(data[col], numOfBin, labels=range(1,numOfBin+1))
    return data
#
t1_4 = pd.DataFrame(t1, columns=['num_lab_procedures'])
binning(t1_4, 5, 'num_lab_procedures')
t1['binned_num_lab_procedures'] = t1_4
print()
t1
```
